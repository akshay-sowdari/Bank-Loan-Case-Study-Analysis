## We need to find driver variables that causes loan default i.e., factors that affect loan repayment


## Variables with most null values cannot be a driving factor hence we need to romve those columns 

```python
null_prev_app_df=pd.DataFrame(round((prev_app_df.isnull().sum()/prev_app_df.shape[0])*100,2)).reset_index()
```

```python
null_prev_app_df.columns=['column_name','null_values_percent']
```

```python
null_prev_app_df
```

```python
null_prev_app_df_10=null_prev_app_df[null_prev_app_df["null_values_percent"]>10]
```

```python
null_prev_app_df_10["column_name"].shape
```

```python
s=null_prev_app_df_10["column_name"]
arr = s.to_numpy()

print(arr)
```

```python
prev_app_df2=prev_app_df
```

```python
prev_app_df2 = prev_app_df2.drop(arr, axis=1)
```

```python
prev_app_df2
```

```python
prev_app_df2.duplicated().sum()
```

```python
prev_app_df_clean = prev_app_df2.dropna()

```
